```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
df = pd.read_csv(url)
df
```

```python
missing_values_count = df.isnull().sum()

# Find the column with the most missing values
column_with_most_missing_values = missing_values_count.idxmax()

# Print the column name with the most missing values
print("Column with the most missing values:", column_with_most_missing_values)
```

```python
nan_counts = df.isnull().sum()  # Count NaN values in each column
nan_counts = nan_counts[nan_counts > 0]  # Filter columns with NaN values

# Plotting NaN values
plt.figure(figsize=(10, 6))
nan_counts.plot(kind='bar')
plt.title('NaN Values in DataFrame')
plt.xlabel('Columns')
plt.ylabel('Count of NaN Values')
plt.show()
```
